```python
import pandas as pd
import numpy as np
import re
import pycountry_convert as pc
```

- Bộ dữ liệu : Road traffic injuries
- Đơn vị tính: Estimated road traffic death rate (per 100 000 population)

```python
data = pd.read_csv("Raw_data/RS_198.csv", header=[0, 1])

def extract_first_number(cell):
    match = re.match(r"([\d\.]+)", str(cell))  
    return float(match.group(1)) if match else cell  

data.iloc[:, 1:] = data.iloc[:, 1:].applymap(extract_first_number)

data.head()
```

```python
continent_manual = {
    "Bolivia (Plurinational State of)": "South America",
    "Cote d'Ivoire": "Africa",
    "Iran (Islamic Republic of)": "Asia",
    "Micronesia (Federated States of)": "Oceania",
    "Netherlands (Kingdom of the)": "Europe",
    "Republic of Korea": "Asia",
    "Turkiye": "Asia",
    "Timor-Leste": "Asia",
    "Venezuela (Bolivarian Republic of)": "South America"
}

def get_continent(country):
    try:
        country_code = pc.country_name_to_country_alpha2(country)  
        continent_code = pc.country_alpha2_to_continent_code(country_code)  
        continent_mapping = {
            "AF": "Africa",
            "AS": "Asia",
            "EU": "Europe",
            "NA": "North America",
            "SA": "South America",
            "OC": "Oceania",
            "AN": "Antarctica"
        }
        item = continent_mapping.get(continent_code, "Unknown")
        if(item == "Unknown"):
            return continent_manual[country]

        return item
    except:
        return continent_manual[country]

country_col = ('Year', 'Countries, territories and areas')

data[('', 'Continent')] = data[country_col].apply(get_continent)



data.to_csv("Preprocessing_data/data.csv", index=False)
data
```
